# Implementing the Perceptron

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
image_path = 'weight_update_eq.PNG'  # Update this path to your image file
image = Image.open(image_path)

# Display the image
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
import numpy as np
#set the seed for reproducibility
np.random.seed(42)

# Define the input-output pairs for the AND gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.random.randn(2)  # Random weights for 2 inputs
bias = 0

# Define the    activation function (step function)
def step_function(x):
    return 1 if x >= 0 else 0

#For all negative values it should return 0 and for all positive values it should return 1
#This is a quick way to check if any function is working correctly
assert step_function(0) == 1
assert step_function(1) == 1
assert step_function(-1) == 0
assert step_function(-11111111111112) == 0

#Training loop

learning_rate=0.1
for epoch in range(10):
    print("--------------------------------")
    print("Epoch", epoch+1)
    total_error=0
    for i in range(len(inputs)):
        print("inputs",inputs[i])
        print("Raw weights",weights)
        #Calculate the dot product of the input and weights
        dot_product = np.dot(inputs[i], weights)
        
        print("dot_product",dot_product)
        #Apply the step function to the dot product
        output = step_function(dot_product + bias)
        print("output",output)
        #Update the weights and bias
        #compute error
        error = outputs[i] - output
        total_error+=error
        print("error for this iteration",error)
        print("total error for this epoch",total_error)
        #Update the weights and bias
        weights=weights+learning_rate*error*inputs[i] 
        bias=bias+learning_rate*error
        print("weights after update",weights)
        print("bias after update",bias)
        print("--------------------------------")

    if total_error==0:
        print("Converged in",epoch+1,"epochs")
        break
```

# The perceptron is a foundational model in machine learning and neural networks, but it has several limitations. Here are some of the key limitations of the perceptron:

## 1. Linearly Separable Data Only
- **Limitation:** The perceptron can only classify linearly separable data. This means that it can only find a linear decision boundary (a straight line in 2D, a hyperplane in higher dimensions) to separate different classes.
- **Example:** The perceptron can successfully classify data points for logical operations like AND and OR, but it fails for operations like XOR (exclusive OR), where the classes are not linearly separable.

## 2. Single Layer Structure
- **Limitation:** A perceptron consists of a single layer of output nodes (no hidden layers). This restricts its ability to learn complex patterns and relationships in the data.
- **Example:** More complex functions or patterns require multiple layers (hidden layers) to capture non-linear relationships, which a single-layer perceptron cannot do.

## 3. Binary Classification
- **Limitation:** The basic perceptron is designed for binary classification tasks. While it can be extended to multi-class classification using techniques like one-vs-all, it is not inherently designed for this purpose.
- **Example:** For multi-class problems, more advanced architectures like softmax regression or multi-layer neural networks are typically used.

## 4. Sensitivity to Learning Rate
- **Limitation:** The choice of learning rate () can significantly affect the training process. If the learning rate is too high, the perceptron may fail to converge or oscillate around the optimal solution. If it is too low, the training process can be very slow.
- **Example:** Finding the right learning rate often requires experimentation and can be a source of instability in training.

## 5. Noisy Data
- **Limitation:** The perceptron is sensitive to noisy data and outliers. A few misclassified points can significantly affect the weights and lead to poor generalization.
- **Example:** In datasets with a lot of noise, the perceptron may struggle to find a good decision boundary, leading to overfitting or underfitting.

## 6. Lack of Hidden Layers
- **Limitation:** Without hidden layers, the perceptron cannot learn hierarchical features or representations of the input data. This limits its ability to model complex functions.
- **Example:** In image recognition tasks, features like edges, shapes, and textures often require multiple layers to be effectively learned.

## Conclusion
While the perceptron laid the groundwork for neural networks and machine learning, its limitations led to the development of more advanced models, such as multi-layer perceptrons (MLPs) and deep neural networks (DNNs), which can handle non-linear separability, multi-class classification, and complex feature hierarchies.
